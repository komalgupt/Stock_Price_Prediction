Engineering feature for Time seires dataset : 1. technical indicator like MACD and RSI 2. Lag feature like past closing price and valatility measures like standard deviation

```python
#Momentum : calculate various momentum indicator . these indictors help assess the speed and direction of price movement in financial instrument.
#volatiliy :  calculate volatility measures. It gauges the magnitude of price flutuations over time. (standard deviation)
from ta import momentum, volatility
```

```python
#converting Date to Datetime formate :  Timestamps are crucial for recording the exact time of trades, orders, or other financial activities.
df_new['Date'] = pd.to_datetime(df_new['Date'])
```

```python
df_new.head()
```

```python
#Engineering features
# 1. teachnical indicatoes (using 'ta' library)
#Relative Strength Index (RSI): Measures the relative strength or weakness of a stock by comparing recent gains to recent losses. High RSI values might indicate overbought conditions (potential for a price decline), while low RSI values might suggest oversold conditions (potential for a price increase).

df_new['RSI'] = momentum.rsi(df_new['Adj Close'])
```

```python
df_new['RSI']
```

```python
#calculate Moving Average Convergence Divergence (MACD): Compares two moving averages of the price to identify potential trend reversals based on the convergence or divergence of the averages.
#The exponential moving average (EMA meaning) is a technical indicator that determines the direction in which the price of a security is moving based on past prices.
slow_ema = df_new['Adj Close'].ewm(span = 26, min_periods=26).mean()
fast_ema = df_new['Adj Close'].ewm(span = 12, min_periods=12).mean()
df_new['MACD'] = fast_ema - slow_ema
df_new['MACD_signal'] = df_new['MACD'].ewm(span = 9, min_periods=9).mean()
df_new['MACD_diff'] = df_new['MACD']-df_new['MACD_signal']

```

```python
df_new['MACD']
```

```python
df_new['MACD_diff']
```

```python
#Lag features (past values)
#Standard Deviation: Captures the dispersion of data points around the average price. A higher standard deviation indicates larger price swings.

df_new['Previous_close'] = df_new['Adj Close'].shift(1)
window = 5
df_new['Past_volatility'] = df_new['Adj Close'].rolling(window = window).std
```
